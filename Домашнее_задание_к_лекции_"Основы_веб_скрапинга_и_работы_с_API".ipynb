#### Обязательная часть
Будем парсить страницу со свежеми новостям на habr.com/ru/all/

Вам необходимо собирать только те статьи, в заголовке которых встречается хотя бы одно требуемое ключевое слово. Эти слова определяем в начале кода в переменной, например:

KEYWORDS = ['python', 'парсинг']

В итоге должен формироваться датафрейм вида: <дата> - <заголовок> - <ссылка>

```python
import requests
from bs4 import BeautifulSoup
```

```python
KEYWORDS = ['python', 'парсинг']
```

```python
# получаем страницу со статьями
url = 'https://habr.com/ru/all/'
req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
# если нужны все страницы сразу
pagination = soup.find_all('div', class_ = 'tm-pagination__page-group')
pagination[1].text.strip() # последняя страница
```

```python
# извлекаем список статей на одной странице
articles = soup.find_all('article', class_ = 'tm-articles-list__item')
articles
```

```python
# добираемся до заголовка статьи и ссылки на нее
article_titles_links = list(map(lambda x: x.find('a', class_ = 'tm-article-snippet__title-link'), articles))
article_titles_links
```

```python
# извлекаем ссылки
article_links = list(map(lambda x: x.get('href'), article_titles_links))
article_links
```

```python
# формируем полноценные ссылки
article_all_links = list(map(lambda x: 'https://habr.com' + x, article_links))
article_all_links
```

```python
# извлекаем заголовки
article_titles = list(map(lambda x: x.find('span').text, article_titles_links))
article_titles
```

```python
import time
```

```python
# возвращает список статей до указанной страницы включая
def get_article_list(pages):
  articles_list = []
  for page in range(1, pages+1):
    req = requests.get('https://habr.com/ru/all/page' + str(page))
    time.sleep(0.3)
    soup = BeautifulSoup(req.text)
    articles = soup.find_all('article', class_ = 'tm-articles-list__item')
    articles_list += articles
  return articles_list

article_list = get_article_list(10)
article_list[0]
```

```python
# возвращает заголовок статьи (если возращается None, то это не статья)
def get_article_title(article):
  article_title_link = article.find('a', class_ = 'tm-article-snippet__title-link')
  if article_title_link != None:
    return article_title_link.find('span').text
  return None

get_article_title(article_list[0])
```

```python
# возвращает полную ссылку на статью (если возращается None, то это не статья)
def get_article_link(article):
  if get_article_title(article) != None:
    article_link = article.find('a', class_ = 'tm-article-snippet__title-link').get('href')
    return 'https://habr.com' + article_link
  return None

get_article_link(article_list[0])
```

```python
# возвращает список нужных нам статей
def get_desired_articles(article_list, KEYWORDS):
  desired_articles = []
  for article in article_list:
    if get_article_title(article) != None:
      if any([keyword in get_article_title(article).lower() for keyword in KEYWORDS]):
        desired_articles.append(article)
  return desired_articles

get_desired_articles(article_list, KEYWORDS)
```

```python
import pandas as pd
```

```python
# формируем датафрейм
def get_habr_articles_df(article_list, KEYWORDS):
  habr_articles_df = pd.DataFrame()
  desired_articles = get_desired_articles(article_list, KEYWORDS)
  for article in desired_articles:
    date = pd.to_datetime(article.find('time').get('datetime'), dayfirst=True).date()
    title = get_article_title(article)
    link = get_article_link(article)
    row = {'дата': date, 'заголовок': title, 'ссылка': link}
    habr_articles_df = habr_articles_df.append(row, ignore_index=True)
  return habr_articles_df
```

```python
article_list = get_article_list(10)
habr_articles_df = get_habr_articles_df(article_list, KEYWORDS)
habr_articles_df
```

```python
# возращает текст статьи
def get_article_contents(article):
  req = requests.get(get_article_link(article))
  soup = BeautifulSoup(req.text)
  article_contents = soup.find('div', class_ = 'tm-article-body').find_all('p')
  return ' '.join([block.text for block in article_contents])

get_article_contents(article_list[0])
```

```python
# возвращает список нужных нам статей (поиск ключевых слов идет и по заголовкам, и по тексту статьи)
def get_desired_articles_upgrade(article_list, KEYWORDS):
  desired_articles = []
  for article in article_list:
    title = get_article_title(article).lower()
    text = get_article_contents(article).lower()
    if title != None:
      if any([keyword in title or keyword in text for keyword in KEYWORDS]):
        desired_articles.append(article)
  return desired_articles

len(get_desired_articles_upgrade(article_list, KEYWORDS))
```

```python
# формируем датафрейм
def get_habr_articles_upgrade_df(article_list, KEYWORDS):
  habr_articles_upgrade_df = pd.DataFrame()
  desired_articles = get_desired_articles_upgrade(article_list, KEYWORDS)
  for article in desired_articles:
    date = pd.to_datetime(article.find('time').get('datetime'), dayfirst=True).date()
    title = get_article_title(article)
    link = get_article_link(article)
    text = get_article_contents(article)
    row = {'дата': date, 'заголовок': title, 'ссылка': link, 'текст_статьи': text}
    habr_articles_upgrade_df = habr_articles_upgrade_df.append(row, ignore_index=True)
  return habr_articles_upgrade_df
```

```python
article_list = get_article_list(10)
get_habr_articles_upgrade_df(article_list, KEYWORDS)
```

```python
# обернем все в функцию, принимающую короткий адрес группы, число постов и возвращающую датафрейм
def get_wall_posts_df(group_name, posts_number):
  wall_request = 'https://api.vk.com/method/wall.get'
  params = {
    'domain': group_name,
    'count': posts_number,
    'access_token': '',
    'v': '5.131'
  }
  res = requests.get(wall_request, params)

  wall_posts_df = pd.DataFrame()

  for post in res.json()['response']['items']:
    date = pd.to_datetime(post['date'], unit='s')
    text = post['text']
    row = {'дата поста': date, 'текст поста': text}
    wall_posts_df = wall_posts_df.append(row, ignore_index=True)

  return wall_posts_df

get_wall_posts_df('netology', 50)
```
