### Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
def classification(rating):
  if rating <= 2:
    return 'Низкий рейтинг'
  elif 2 < rating <= 4:
    return 'Средний рейтинг'
  elif 4 < rating <= 5:
    return 'Высокий рейтинг'
  else:
    return 'Нет рейтинга'
```

```python
ratings['class'] = ratings['rating'].apply(classification)
ratings
```

### Задание 2
Используем файл keywords.csv.

Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.

Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

Результат классификации запишите в отдельный столбец region.

```python
keywords = pd.read_csv('ml-latest-small/keywords.csv')
keywords.head(10)
```

```python
def geo_class(keyword):
  for region, cities in geo_data.items():
    for city in cities:
      if city in keyword:
        return region
  return 'undefined' 
```

```python
keywords['region'] = keywords['keyword'].apply(geo_class)
keywords[ keywords['region'] != 'undefined'].head(10)
```

```python
keywords[ keywords['keyword'].str.contains('петербург')].head(10)
```

```python
keywords[ keywords['keyword'].str.contains('сахалин')].head(10)
```

### Задание 3 (бонусное)
Есть мнение, что “раньше снимали настоящее кино, не то что сейчас”. Ваша задача проверить это утверждение, используя файлы с рейтингами фильмов из прошлого домашнего занятия (файл movies.csv и ratings.csv из базы grouplens.org.../movielens). Т. е. проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

При этом мы не будем затрагивать субьективные факторы выставления этих рейтингов, а пройдемся по следующему алгоритму:

1. В переменную years запишите список из всех годов с 1950 по 2010.
2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:
  1. для каждой строки пройдите по всем годам списка years
  2. если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска
  3. если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год
3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’
4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
years = list(range(1950, 2011))
print(years)
```

```python
def production_year(title):
  for year in years:
    if str(year) in title:
      return year
  return 1900
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
movies.head()
```

```python
movies['year'] = movies.title.apply(production_year)
movies.tail(10)
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings
```

```python
ratings_merged = ratings[['movieId', 'rating']].merge(movies[['movieId', 'title', 'year']], how='left', on='movieId')
ratings_merged
```

```python
ratings_merged.groupby('year').mean().sort_values('rating', ascending=False)['rating'].head(20)
```
